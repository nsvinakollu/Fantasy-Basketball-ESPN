```python
import pandas as pd
import qgrid as qg
```

```python
espnTeams = pd.read_csv("Week_1_2021.csv")
bballRef = pd.read_csv("Player_Stats_Week_3.csv")
bballRefAdvanced = pd.read_csv("Player_Advanced_Stats_Week_3.csv")
```

```python
bballRef = bballRef.loc[:, ~bballRef.columns.str.contains("^Unnamed")]
espnTeams = espnTeams.loc[:, ~espnTeams.columns.str.contains("^Unnamed")]
bballRefAdvanced = bballRefAdvanced.loc[
    :, ~bballRefAdvanced.columns.str.contains("^Unnamed")
]

charsReplace = ["-", ".", " ", "III", "IV", "II"]
for chars in charsReplace:
    espnTeams["Player"] = espnTeams["Player"].str.replace(chars, "")
    bballRef["Player"] = bballRef["Player"].str.replace(chars, "")
    bballRefAdvanced["Player"] = bballRefAdvanced["Player"].str.replace(chars, "")
espnTeams["Player"] = espnTeams["Player"].str.lower()
bballRef["Player"] = bballRef["Player"].str.lower()
bballRefAdvanced["Player"] = bballRefAdvanced["Player"].str.lower()


regularStats = espnTeams.merge(bballRef)
allStats = bballRef.merge(bballRefAdvanced, on = ["Player", "Age", "Year"])
teamStats = espnTeams.merge(allStats, on = "Player")
# teamStats = teamStats.merge(bballRefAdvanced)

teamStats
allStats
```

```python
bballRef
bballRef = bballRef[bballRef['Player'].notna()]

qg_widget = qg.show_grid(bballRef, show_toolbar=True)
qg_widget

```

```python
fantasyScoreAvg = pd.DataFrame(bballRef, columns = ['Player','Tm', 'Pos', 'G', 'GS', 'MP'])
fantasyScore = pd.DataFrame()

# Stats to Fantasy Value Conversion
fgm = .5
fga = -.5
ftm = .25
fta = -.25
oreb = 1.25
dreb = 1
ast = 1.5
stl = 2.75
blk = 2.75
to = -1
tf = -1
pts = 1


fantasyScore['FG'] = bballRef['FG'] * fgm
fantasyScore['FGA'] = bballRef['FGA'] * fga
fantasyScore['FTM'] = bballRef['FT'] * ftm
fantasyScore['FTA'] = bballRef['FTA'] * fta
fantasyScore['OREB'] = bballRef['ORB'] * oreb
fantasyScore['DREB'] = bballRef['DRB'] * dreb
fantasyScore['AST'] = bballRef['AST'] * ast
fantasyScore['STL'] = bballRef['STL'] * stl
fantasyScore['BLK'] = bballRef['BLK'] * blk
fantasyScore['TO'] = bballRef['TOV'] * to
fantasyScore['PTS'] = bballRef['PTS'] * pts

fantasyScore['Avg. Total'] = fantasyScore.sum(axis=1)
fantasyScore['Total'] = fantasyScore['Avg. Total'] * bballRef['G']

# qg_widget = qg.show_grid(fantasyScore, show_toolbar=True)
# qg_widget

fantasyScoreAvg = fantasyScoreAvg.join(fantasyScore)
fantasyScoreAvg
qg_widget = qg.show_grid(fantasyScoreAvg, show_toolbar=True)
qg_widget
```
