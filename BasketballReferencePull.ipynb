```python
import requests
import pandas as pd
import json as js
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Player Stats for 2021
year = "2021"

url = "https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_per_game.html"

html = urlopen(url)

soup = BeautifulSoup(html)

# use findALL() to get the column headers
soup.findAll("tr", limit=1)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll("tr", limit=2)[0].findAll("th")]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers = headers[1:]
headers.insert(0, "Year")

# avoid the first header row
rows = soup.findAll("tr")[1:]
player_stats = [
    (td.getText() for td in rows[i].findAll("td")) for i in range(len(rows))
]

yearDF = pd.DataFrame([year] * len(rows))
playerStats = pd.DataFrame(player_stats)
playerStatsDF = pd.concat([yearDF, playerStats], axis=1)
playerStatsDF.columns = headers
```

```python
playerStatsDF.to_csv('Player_Stats_Week_3.csv')
```

```python
# Advanced Stats for 2021
url = "https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_advanced.html"

html = urlopen(url)

soup = BeautifulSoup(html)

# use findALL() to get the column headers
soup.findAll("tr", limit=1)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll("tr", limit=2)[0].findAll("th")]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers = headers[1:]  # avoid the first header row
headers.insert(0, "Year")

rows = soup.findAll("tr")[1:]
player_advanced_stats = [
    [td.getText() for td in rows[i].findAll("td")] for i in range(len(rows))
]

yearDF = pd.DataFrame([year] * len(rows))
playerAdvancedStats = pd.DataFrame(player_advanced_stats)
playerAdvancedStatsDF = pd.concat([yearDF, playerAdvancedStats], axis=1)
playerAdvancedStatsDF.columns = headers

playerAdvancedStatsDF
```

```python
playerAdvancedStatsDF.to_csv('Player_Advanced_Stats_Week_3.csv')
```

```python
#Stats for Multiple Years
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021']

playerStatsYearsDF = pd.DataFrame()

for year in years:

    url = "https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_per_game.html"

    html = urlopen(url)

    soup = BeautifulSoup(html)

    # use findALL() to get the column headers
    soup.findAll("tr", limit=1)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll("tr", limit=2)[0].findAll("th")]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers.insert(0, "Year")

    # avoid the first header row
    rows = soup.findAll("tr")[1:]
    player_stats = [
        (td.getText() for td in rows[i].findAll("td")) for i in range(len(rows))
    ]

    yearDF = pd.DataFrame([year] * len(rows))
    playerStats = pd.DataFrame(player_stats)
    playerStatsDF = pd.concat([yearDF, playerStats], axis=1)
    playerStatsYearsDF = pd.concat([playerStatsYearsDF, playerStatsDF])
playerStatsYearsDF.columns = headers
playerStatsYearsDF
```
